```python
def parse_button(button, lamp_count):
    assert button[0] == '('
    assert button[-1] == ')'
    button = [int(v) for v in button[1:-1].split(",")]
    mask = [0]*lamp_count
    for b in button:
        mask[b] = 1
    return mask
def parse_joltage(joltage):
    assert joltage[0] == '{'
    assert joltage[-1] == '}'
    joltage = [int(v) for v in joltage[1:-1].split(",")]
    return joltage
def parse_line(line):
    line = line.split()[1:]
    joltage = parse_joltage(line[-1])
    lamp_count = len(joltage)
    buttons = [parse_button(b, lamp_count) for b in line[:-1]]
    return (buttons, joltage)
with open("input.txt") as f:
    inp = [parse_line(l) for l in f]
print(inp)
```

```python
def solve_line(line):
    (buttons,joltage) = line
    lamp_count = len(joltage)
    p = MixedIntegerLinearProgram(maximization=False, solver='GLPK')
    w = p.new_variable(integer=True, nonnegative=True)
    acc = [0]*lamp_count
    for i in range(lamp_count):
        p.add_constraint((matrix([a[i] for a in buttons]) * w) == joltage[i])
    p.set_objective(sum([w[i] for i in range(len(buttons))]))
    return p.solve()
sum([solve_line(line) for line in inp])
```
